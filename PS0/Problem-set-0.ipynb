### 1(a)

Note that

$$\begin{align*}
b^Tx=
\begin{bmatrix}b_1&\dots&b_n\end{bmatrix}\begin{bmatrix}x_1\\\vdots\\x_n\end{bmatrix}=\sum_{i=1}^nb_ix_i\end{align*}$$

Hence

$$\begin{align*}
\frac{\partial b^Tx}{\partial x_i}=b_i\end{align*}$$

And

$$\nabla_x (b^Tx)=\begin{bmatrix}\frac{\partial b^Tx}{\partial x_1}\\\vdots\\\frac{\partial b^Tx}{\partial x_n}\end{bmatrix}=\begin{bmatrix}b_1\\\vdots\\b_n\end{bmatrix} = b$$

Similarly for $x^TAx$:

$$\begin{align*}
x^TAx &= x^T(Ax) \\
&= x^T
\begin{bmatrix}
 A_1 x \\ 
 A_2 x \\
 \cdots \\
 A_n x
\end{bmatrix} \\
&= \begin{bmatrix}x_{1}&\dots&x_{n}\end{bmatrix}
\begin{bmatrix}
 \sum_{j=1}^{n} A_{1j} x_j \\ 
 \sum_{j=1}^{n} A_{2j} x_j \\
 \cdots \\
 \sum_{j=1}^{n} A_{nj} x_j
\end{bmatrix} \\
&= x_1\sum_{j=1}^{n} A_{1j} x_j + x_2\sum_{j=1}^{n} A_{2j} x_j + \cdots + x_n \sum_{j=1}^{n} A_{nj} x_j \\
&= \sum_{i=1}^n \sum_{j=1}^{n} x_i A_{ij} x_j \\
&= \sum_{i=1}^n \sum_{j=1}^{n} A_{ij} x_i x_j
\end{align*}$$

Next, in the second equality, we split out the double sum into the three disjoint cases: (1) neither $i$ nor $j$ equals $k$, (2) exactly one of $i$ or $j$ equals $k$, and (3) both equal $k$:

$$\begin{align*}
\frac{\partial x^TAx}{\partial x_k}&=\frac{\partial \sum_{j=1}^n \sum_{i=1}^{n} A_{ij} x_i x_j}{\partial x_k}\\
&=\frac{\partial \big(\sum_{j=1,j\neq k}^n \sum_{i=1,i\neq k}^{n} A_{ij} x_i x_j+\sum_{i=k}^k\sum_{j=1,j\neq k}^{n} A_{ij} x_i x_j+\sum_{j=k}^k\sum_{i=1,i\neq k}^{n} A_{ij} x_i x_j+A_{kk}x_kx_k\big)}{\partial x_k}\\
&=\frac{\partial \big(\sum_{j=1,j\neq k}^n \sum_{i=1,i\neq k}^{n} A_{ij} x_i x_j+\sum_{j=1,j\neq k}^{n} A_{kj} x_k x_j+\sum_{i=1,i\neq k}^{n} A_{ik} x_i x_k+A_{kk}x_k^2\big)}{\partial x_k}\\
&=\frac{\partial \big(\sum_{j=1,j\neq k}^{n} A_{kj} x_k x_j+\sum_{i=1,i\neq k}^{n} A_{ik} x_i x_k+A_{kk}x_k^2\big)}{\partial x_k}\\
&=\frac{\partial \big(\sum_{j=1,j\neq k}^{n} A_{kj} x_k x_j+\sum_{i=1,i\neq k}^{n} A_{ki} x_k x_i+A_{kk}x_k^2\big)}{\partial x_k}\tag{A symmetric}\\
&=\frac{\partial \big(2\sum_{j=1,j\neq k}^{n} A_{kj} x_k x_j+A_{kk}x_k^2\big)}{\partial x_k}\\
&=\frac{\partial \big(2\sum_{j=1,j\neq k}^{n} A_{kj} x_k x_j\big)}{\partial x_k}+\frac{\partial A_{kk}x_k^2}{\partial x_k}\\
&=2\sum_{j=1,j\neq k}^{n}\frac{\partial A_{kj} x_k x_j}{\partial x_k}+2A_{kk}x_k\\
&= 2 \sum_{j = 1,j\neq k}^{n} A_{kj}x_j+2A_{kk}x_k \\
&= 2 \sum_{j = 1}^{n} A_{kj}x_j \\
&= 2 \begin{bmatrix}A_{k1}&\dots&A_{kn}\end{bmatrix}\begin{bmatrix}x_1\\\vdots\\x_n\end{bmatrix} \\
&= 2 A_k x \\
\end{align*}$$

Hence the gradient is

$$\begin{align*}
\nabla_x (x^TAx) = \begin{bmatrix}\frac{\partial x^T(Ax)}{\partial x_1}\\\vdots\\\frac{\partial x^T(Ax)}{\partial x_n}\end{bmatrix}
= 2 \begin{bmatrix}A_1 x\\\vdots\\A_n x
\end{bmatrix} = 2Ax
\end{align*}$$

And the answer to the question is

$$\nabla_x f(x) = \nabla_x \Big(\frac{1}{2} x^T A x + b^T x\Big) = \frac{1}{2}\nabla_xx^TAx + \nabla_xb^Tx = Ax + b$$,

which is analogous to the univariate case:

$$\frac{d(\frac{1}{2}ax^2 + bx)}{dx} = ax + b$$.

### 1(b)

The chain rule gives

$$\begin{align*}
\frac{\partial f(x)}{\partial x_i}=\frac{\partial g(h(x))}{\partial x_i}=g'(h(x))\frac{\partial h(x)}{\partial x_i}
\end{align*}$$

Hence

$$\begin{align*}
\nabla_xf(x) = \begin{bmatrix}\frac{\partial f(x)}{\partial x_1}\\\vdots\\\frac{\partial f(x)}{\partial x_n}\end{bmatrix}= \begin{bmatrix}g'(h(x))\frac{\partial h(x)}{\partial x_1}\\\vdots\\g'(h(x)) \frac{\partial h(x)}{\partial x_n} \\\end{bmatrix}=g'(h(x))\begin{bmatrix}\frac{\partial h(x)}{\partial x_1}\\\vdots\\ \frac{\partial h(x)}{\partial x_n} \\\end{bmatrix}=g'(h(x))\nabla_xh(x)
\end{align*}$$

### 1(c)

Part (a) gives

$$
\frac{\partial^2 b^Tx}{\partial x_k \partial x_l}=\frac{\partial b_l}{\partial x_k}=0
$$

So the Hessian is

$$
\nabla_x^2b^Tx=\begin{bmatrix}\frac{\partial^2 b^Tx}{\partial x_1 \partial x_1}&\dots&\frac{\partial^2 b^Tx}{\partial x_1 \partial x_n}\\\vdots&\ddots&\vdots\\\frac{\partial^2 b^Tx}{\partial x_n \partial x_1}&\dots&\frac{\partial^2 b^Tx}{\partial x_n \partial x_n}\end{bmatrix}=\begin{bmatrix}0&\dots&0\\\vdots&\ddots&\vdots\\0&\dots&0\end{bmatrix}=0
$$

Part (a) gives

$$\begin{align*}
\frac{\partial x^TAx}{\partial x_k}=2A_k x=2 \begin{bmatrix}A_{k1}&\dots&A_{kn}\end{bmatrix}\begin{bmatrix}x_1\\\vdots\\x_n\end{bmatrix}
=2\sum_{j = 1}^{n} A_{kj}x_j \\
\end{align*}$$

Hence $x^TAx$ is twice continuously differentible and

$$\begin{align*}
\frac{\partial^2 x^TAx}{\partial x_k\partial x_l}=\frac{\partial^2 x^TAx}{\partial x_l\partial x_k}=2\frac{\partial\sum_{j = 1}^{n} A_{kj}x_j}{\partial x_l}=2A_{kl} \\
\end{align*}$$

So the Hessian is

$$
\nabla_x^2x^TAx=\begin{bmatrix}\frac{\partial^2 x^TAx}{\partial x_1 \partial x_1}&\dots&\frac{\partial^2 x^TAx}{\partial x_1 \partial x_n}\\\vdots&\ddots&\vdots\\\frac{\partial^2 x^TAx}{\partial x_n \partial x_1}&\dots&\frac{\partial^2 x^TAx}{\partial x_n \partial x_n}\end{bmatrix}=\begin{bmatrix}2A_{11}&\dots&2A_{1n}\\\vdots&\ddots&\vdots\\2A_{n1}&\dots&2A_{nn}\end{bmatrix}=2\begin{bmatrix}A_{11}&\dots&A_{1n}\\\vdots&\ddots&\vdots\\A_{n1}&\dots&A_{nn}\end{bmatrix}=2A
$$

and the solution is

$$
\nabla_x^2 f(x) = \nabla_x^2 \Big(\frac{1}{2}x^T A x + b^Tx\Big)=\frac{1}{2}2A=A
$$

which is analogous to

$$\frac{d^2(\frac{1}{2}ax^2 + bx)}{dx^2} = a$$.

__Corrolary:__ $w(x)=x^Tx$ is convex.

__Proof__ Recall that $w$ is convex if and only if its Hessian is positive semi-definite. Note that

$$
\nabla_x^2w(x)=\nabla_x^2x^TIx=2I
$$

and for any vector $v\in\mathbb{R}^n$

$$
v^T2Iv=2v^Tv=2\lvert v\rvert^2=2\sum_iv_i^2>0
$$

$\blacksquare$

__Corrolary:__ Affine functions are convex.

__Proof__ For $a\in\mathbb{R}$, let $q(x)=b^Tx+a$. Note that

$$
\nabla_x^2q(x)=\nabla_x^2(b^Tx)+\nabla_x^2(a)=0_{n\times n}
$$

and for any vector $v\in\mathbb{R}^n$

$$
v^T0v=0
$$

$\blacksquare$

### 1(d)

Define $h(x)=a^Tx$ so that $h:\mathbb{R}^n\mapsto\mathbb{R}$ and $f(x)=g\circ h(x)$. Then part (a) gives $\nabla_xh(x)=a$ and part (b) gives

$$\nabla_xf(x)=g'(h(x))\nabla_xh(x) = g'(a^Tx) \cdot a$$

Part (b) also gives the partials:

$$\begin{align*}
\frac{\partial g\circ h(x)}{\partial x_i}=g'(h(x))\frac{\partial h(x)}{\partial x_i}=g'(a^Tx)a_i
\end{align*}$$

Since $f=g\circ h$ is twice continuously differentiable, then the second partials are

$$\begin{align*}
\frac{\partial^2 f(x)}{\partial x_i\partial x_j}=\frac{\partial^2 g\circ h(x)}{\partial x_j\partial x_i}=\frac{\partial g'(a^Tx)a_i}{\partial x_j}=a_i\frac{\partial g'(\sum_{k=1}^na_kx_k)}{\partial x_j}=a_ia_jg''(a^Tx)
\end{align*}$$

And the Hessian is

$$\begin{align*}
\nabla_x^2f(x)=\begin{bmatrix}a_1a_1g''(a^Tx)&\dots&a_1a_ng''(a^Tx)\\\vdots&\ddots&\vdots\\a_na_1g''(a^Tx)&\dots&a_na_ng''(a^Tx)\end{bmatrix}=g''(a^Tx)\begin{bmatrix}a_1a_1&\dots&a_1a_n\\\vdots&\ddots&\vdots\\a_na_1&\dots&a_na_n\end{bmatrix}=g''(a^Tx)\begin{bmatrix}a_1\\\dots\\a_n\end{bmatrix}\begin{bmatrix}a_1&\dots&a_n\end{bmatrix}=g''(a^Tx)aa^T
\end{align*}$$

\begin{align*}
A &= zz^T \\

\end{bmatrix} \\
 &= R \\
\end{align*}

The $\rightarrow$ means row transformation by multiplying row $1$ with $-\frac{z_i}{z_1}$, and add it to row $i$. This is possible because $z$ is given as a _non-zero n-vector_.

So the rank of $A$ is $1$.

To obtain its nullspace, we need to find all possible $x$ vectors that satisfy $Ax = 0$, Before that, let's consider the dimensions of four fundamental subspaces in linear algebra:


|                | Symbol   | Dimension | Note                                                                                |
|----------------|----------|-----------|-------------------------------------------------------------------------------------|
| Row space      | $C(A^T)$ | 1         | Equal to rank                                                                                    |
| Column space   | $C(A)$   | 1         | Equal to rank                                                                       |
| Nullspace      | $N(A)$   | n - 1     | Usually the dimension of nullspace is $m - 1$, but given A is symmetric, so $m = n$ |
| Left nullspace | $N(A^T)$ | n - 1     | - |

Therefore, the nullspace of $A$ consists of $n-1$ vectors, which could be derived from $n-1$ free variables. Result:

\begin{align*}
Ax &= 0 \\
Rx &= 0 \\
z_1 \sum_{i=1}^{n} z_i x_i & = 0 \\
\sum_{i=1}^{n} z_i x_i &= 0
\end{align*}

Therefore, by setting any of the $n=1$ free variables to $1$, and the rest to $0$, and then calculating a value for the pivot variable ($x_1$), we can obtain $n-1$ independent vectors for $x$, and by multiplying each vector with an arbitrary constant, we obtain the nullspace, i.e. all possible vectors that satisfy $Ax = 0$.

\begin{align*}
N(A) &=
c_2 \begin{bmatrix}
 -z_2/z_1 \\ 
 1 \\
 0 \\
 0 \\
 \vdots \\
 0
\end{bmatrix} + 
c_3 \begin{bmatrix}
 -z_3/z_1 \\
 0 \\
 1 \\
 0 \\
 \vdots \\
 0
\end{bmatrix} +
\cdots +
c_i \begin{bmatrix}
 -z_i/z_1 \\
 0 \\
 \vdots \\
 1 \\
 \vdots \\
 0
\end{bmatrix} + 
\cdots +
c_n \begin{bmatrix}
 -z_n/z_1 \\
 0 \\
 0 \\
 0 \\
 \vdots \\
 1
\end{bmatrix} \\
&= \sum_{k=2}^n c_k \begin{bmatrix}
 -z_k/z_1 \\
 I\{i=k\} \\
 I\{i=k\} \\
 \vdots \\
 I\{i=k\}
\end{bmatrix}
\end{align*}

where $i$ is the row index, $I$ is an indicator function, and $c_k$ is a constant. Note, $k$ starts from $2$ instead of $1$. This is meaningful only when $z_1$ is non-zero.

### 2(c)

Given 

1. $A \in \mathbb{R}^{n\times n}$ is PSD, so $A = A^T$, and $x^T A x \ge 0$, and 
1. $B \in \mathbb{R}^{m\times n}$

Let's denote $C = BAB^T$. We need to prove two parts in order to show that $C$ is PSD, too.

Part 1.

\begin{align*}
C^T &= (BAB^T)^T \\
    &= (B^T)^T A^T B^T \\
    &= B A^T B^T \\
    &= B A B^T \\
    &= C
\end{align*}

Part 2.
\begin{align*}
x^T C x &= x^T (BAB^T) x \\
        &= (x^T B) A (B^T x) \\
        &= (B^T x)^T A (B^T x) \\
        &= y^T A y \\
        &\ge 0
\end{align*}

Here we transformed $x^T C x$ into one that leverages the property of A, $x^T A x \ge 0$, by setting $y = B^T x$. Therefore, $BAB^T$ is PSD.

### 3(a)

Given $A = T \Lambda T^{-1}$, so

\begin{align*}
            & A T = T \Lambda \\
\rightarrow & A[t^{(1)} \cdots t^{(n)}] = [t^{(1)} \cdots t^{(n)}] \Lambda \\
\rightarrow & At^{(i)} = t^{(i)} \lambda_i \\
\end{align*}

The last step is because $\Lambda$ is a diagonal matrix with all off-diagonal values being zeros. Therefore, 

$$ At^{(i)} = \lambda_i t^{(i)} $$

$(t^{(i)}, \lambda_i)$ is an eigenvalue/eigenvector pair.

### 3(b)

##### **Spectral theorem (important!)**: 

If $A \in \mathbb{R}^{n \times n}$ is **symetric** (i.e. $A = A^T$), then $A$ is **diagonalizable** ($\Lambda$) by a **real** **orthogonal** ($U$) matrix. In symbols, 

$$ U^T A U = \Lambda$$

or equivalently,

$$ A = U \Lambda U^T$$


Given $A = A^T$, and $U$ is orthogonal, and $A = U \Lambda U^T$, then

$$AU = U \Lambda U^T U = U \Lambda$$

Following the same logic as in **3(a)**, we get

$$Au^{(i)} = \lambda_i u^{(i)}$$

so $u^{(i)}$ is an eigenvector.
